```python
for_plotting = pd.read_csv('cities_r2.csv')
```

```python
latitude, longitude = [], []
for i in for_plotting['location']:
    latitude.append(float(i.split(',')[0]))
    longitude.append(float(i.split(',')[1]))
```

```python
from mpl_toolkits.basemap import Basemap

fig, ax = plt.subplots(figsize=(20,20))
m = Basemap(resolution='l', # c, l, i, h, f or None
            projection='mill',
            lat_0=20.5937, lon_0=78.9629,
            llcrnrlon=64.31, llcrnrlat= 5.11, urcrnrlon=99.87, urcrnrlat=36.72)


m.fillcontinents(color='0.95')
m.drawcoastlines(color = '0.5', linewidth = 0.25)
m.drawcountries(linewidth = 0.5)
m.drawmapboundary(fill_color= '#e1f7f2')

x,y = m(longitude, latitude)
m.plot(x, y, 'bo', alpha = 0.7)
```
