```python
for_plotting = pd.read_csv('cities_r2.csv')
```

```python
latitude, longitude = [], []
for i in for_plotting['location']:
    latitude.append(float(i.split(',')[0]))
    longitude.append(float(i.split(',')[1]))
```

```python
for_plotting['latitude'] = latitude
for_plotting['longitude'] = longitude
```

```python
import matplotlib as mpl
mpl.rcParams['axes.facecolor'] = 'white'
for_plotting.plot(kind = 'scatter', x = 'longitude', y = 'latitude', figsize = (20, 20), grid = False)
```
